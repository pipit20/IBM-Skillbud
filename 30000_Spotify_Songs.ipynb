```python
# total & persentase NA per kolom (diurutkan desc)
na_col = (df.isna().sum()
          .to_frame('na_count')
          .assign(na_pct=lambda x: (x['na_count']/len(df)).round(4))
          .sort_values('na_count', ascending=False))
na_col.head(20)

# baris dengan NA terbanyak (cek potensi baris “rusak”)
na_row = df.isna().sum(axis=1).sort_values(ascending=False)
na_row.head(10)
```

```python
from langchain_experimental.agents.agent_toolkits.pandas.base import create_pandas_dataframe_agent
from langchain_community.llms import Replicate

llm = Replicate(
    model="ibm-granite/granite-3.3-8b-instruct",
    model_kwargs={"temperature": 0.6, "max_tokens": 512}  # bisa diubah sesuai kebutuhan
)

agent = create_pandas_dataframe_agent(
    llm=llm,
    df=df,
    verbose=True,
    allow_dangerous_code=True
)
```

```python
avg_popularity_by_genre = df.groupby('playlist_genre')['track_popularity'].mean().sort_values(ascending=False)

print(avg_popularity_by_genre)
```
